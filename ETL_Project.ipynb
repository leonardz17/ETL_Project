```python
import json
from pprint import pprint
import requests
from config import eia_key
```

```python
# API call for weekly price of WTI benchmark for crude oil prices in the US

url = "http://api.eia.gov/series/?"

query_url = f"{url}api_key={eia_key}&series_id=PET.RWTC.W"

print(query_url)
```

```python
price_response = requests.get(query_url).json()
print(json.dumps(price_response, indent=4))
```

```python
price_response["series"][0]["data"]
```

```python
# json to df for weekly crude oil prices(WTI US benchmark)

df = pd.DataFrame(price_response["series"][0]["data"])
oil_price_df = df.rename(columns = {0:"Date",1:"Price"})
oil_price_df
```

```python
oil_price_df.dtypes
```

```python
# Convert Date from object to datetime
oil_price_df["Date"] = pd.to_datetime(oil_price_df["Date"])
oil_price_df
```
